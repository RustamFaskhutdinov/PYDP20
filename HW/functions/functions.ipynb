Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
# 1_1
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
# p
def doc_owner(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            print(f"Владелец документа: {document['name']}")
            return
    print("Документ не найден в базе")
# doc_owner()
````

````python
# s    
def find_shelf_by_doc_number(doc_number, print_ = True):
    for shelf, doc_numbers in directories.items():
        if doc_number in doc_numbers:
            if print_:
                if print_: print(f"Документ хранится на полке: {shelf}")
            return shelf
    if print_: print("Документ не найден в базе")
    return
# find_shelf_by_doc_number()
````

````python
# l
def print_all_information():
    for document in documents:
        shelf = find_shelf_by_doc_number(document['number'], print_ = False)
        if shelf is not None:
            print(f"№ {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {shelf}")
# print_all_information()
````

````python
# ads
def add_shelf(shelf):
    if not shelf in directories:
        directories[str(shelf)] = []
        print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories)}")
    else:
        print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories)}")
# add_shelf('2')
````

````python
# ds
def del_shelf(shelf):
    if shelf in directories and not bool(directories[shelf]):
        directories.pop(shelf, None)
        print(f"Полка удалена. Текущий перечень полок: {', '.join(directories)}")
    else:
        print(f"На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories)}")
# del_shelf('3')
````

````python
# d
def del_document_by_number():
    doc_number = input("Введите номер документа:\n")
    for i, document in enumerate(documents[:]):
        if document['number'] == doc_number:
            del(documents[i])
            shelf = find_shelf_by_doc_number(document['number'], print_ = False)
            directories[shelf].remove(doc_number)
            print("Документ удален.")
            print("Текущий список документов:")
            print_all_information()
            return
    print("Документ не найден в базе.")
    print("Текущий список документов:")
    print_all_information()
    
# print(directories, documents, sep = '\n'*3)
# del_document_by_number()
# print(directories, documents, sep = '\n'*3)
````

````python
def move():
    doc_number = input("Введите номер документа:\n")
    new_shelf = input("Введите номер полки:\n")
    if new_shelf not in directories:
        print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}")
        return
    else:
        for document in documents:
            if document['number'] == doc_number:
                current_shelf = find_shelf_by_doc_number(document['number'], print_ = False)
                directories[current_shelf].remove(doc_number)
                directories[new_shelf].append(doc_number)
                print("Документ перемещен.")
                print("Текущий список документов:")
                print_all_information()
                return
        print("Документ не найден в базе.")
        print("Текущий список документов:")
        print_all_information()
        return
# move()
````
