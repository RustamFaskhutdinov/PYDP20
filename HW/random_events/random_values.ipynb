```python
def am_i_lucky(number):
    if number == 777:
        return 200
    elif number == 999:
        return 100
    elif number == 555:
        return 50
    elif number == 333:
        return 15
    elif number == 111:
        return 10
    elif number % 100 == 77:
        return 5
    elif number % 10 == 7:
        return 3
    elif number % 100 == 0 and number != 0:
        return 2
    elif number % 10 == 0:
        return 1
    else:
        return 0
```

```python
%%time
tryes = (10 ** 6) * 5
income = 0
for i in range(tryes):
    income += am_i_lucky(np.random.randint(0,1000))
```
