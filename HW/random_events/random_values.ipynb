```python
import numpy as np
```

Сперва расчитаем теоретический выигрыш.

```python
#777
k = (1 / 1000) * 200

#999
k += (1 / 1000) * 100

#555
k += (1 / 1000) * 50

#333
k += (1 / 1000) * 15

#111
k += (1 / 1000) * 10

#*77
k += (9 / 1000) * 5

#**7
k += (10*9 / 1000) * 3 

#*00
k += (9 / 1000) * 2

#**0
k += ((9*10 + 1) / 1000) * 1 

print(k, round(k-1, 3) , sep='\n')
```

**Теоретический выигрыш с каждой игры в среднем 0.799р что при стоимости игры в 1р дает убыток в 0.201р.**

Далее произведем экспериментальное доказательство этого факта 

```python
def am_i_lucky(number):
    if number == 777:
        return 200
    elif number == 999:
        return 100
    elif number == 555:
        return 50
    elif number == 333:
        return 15
    elif number == 111:
        return 10
    elif number % 100 == 77:
        return 5
    elif number % 10 == 7:
        return 3
    elif number % 100 == 0 and number != 0:
        return 2
    elif number % 10 == 0:
        return 1
    else:
        return 0
```

```python
%%time
tryes = (10 ** 6) * 5
income = 0
for i in range(tryes):
    income += am_i_lucky(np.random.randint(0,1000))
```

```python
income / tryes
```

```python
income / tryes - 1
```

**Практические полученные значения максимально приближены к рассчитанными теоретически.** 

Следовательно можно заключить, что модель дает корректный результат.
