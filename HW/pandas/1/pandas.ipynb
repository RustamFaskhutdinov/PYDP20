## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.<br> Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
rating_data = pd.read_csv('./ml-latest/ratings.csv')
rating_data.head()
```

```python
movies_data = pd.read_csv('./ml-latest/movies.csv')
movies_data.head()
```

```python
rating_data[rating_data['rating'] == 5.]\
.groupby('movieId').count()\
.sort_values(by='userId', ascending = False)\
.merge(movies_data, on='movieId', how='inner').head()
```

```python
del movies_data, rating_data
```

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_data = pd.read_csv('./power_data/power.csv')
power_data.head()
```

```python
power_data = power_data[(power_data['category'].isin([4,12,21])) & (power_data['quantity'] > 0) &\
(power_data['year'].isin(range(2005,2011))) &\
(power_data['country'].isin(['Lithuania','Estonia','Latvia']))]
power_data['quantity'].sum()
```
